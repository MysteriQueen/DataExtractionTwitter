## Data extraction

We extract all relevant raw data for the 3 accounts from twitter using the Tweepy API. 

This section is regarding the raw extraction of data using the API calls available, in order to obtain all relevant data.

In order to obtain the following list, we searched the api documentation for Tweepy (https://docs.tweepy.org/en/stable/api.html) and sought at methods which can acquire useful data of an account.


The list below shows all the raw data that we will be extracting from the accounts:
- Number of accounts that they are following.
- Number of followers.
- Number of original tweets that were posted.
- Number of tweets that the users have liked.
- Number of likes that each original tweet has.
- Number of times that each original tweet have been retweeted.
- Number of retweets that were posted by the user.
- Number of times that the accounts were mentioned.

```python
#Create a list to extract data from all 3 accounts
lUsersData = [api.get_user(screen_name = 'BarackObama'), api.get_user(screen_name = 'iJustine'), api.get_user(screen_name = 'LeoDiCaprio')]

#Go through each user and extract and print data
for oUser in lUsersData:
    sUsername = oUser.name
    iFollowing = oUser.friends_count
    iFollowers = oUser.followers_count
    iTweetsPosted = oUser.statuses_count
    iFavourites = oUser.favourites_count
    print(sUsername, "| Following -", iFollowing, "| Followers -", iFollowers, 
          "| Number of tweets posted -", iTweetsPosted, "|\nNumber of tweets user has liked -", iFavourites, "\n")
    
    #Currently printing all of the tweets
    iCount = 0
    for oTweet in tweepy.Cursor(api.user_timeline, screen_name=oUser.screen_name, tweet_mode="extended").items():
        iCount = iCount + 1
        print(f"{iCount}\t{oTweet.user.id}\t{oTweet.user.screen_name}\t{oTweet.created_at}\t{oTweet.full_text}")


#lTweetsData = api.get_retweets(screen_name = 'BarackObama')
#lTweetsData = tweepy.Cursor(api.user_timeline,id='BarackObama').items()

#for oTweet in lTweetsData:
```
